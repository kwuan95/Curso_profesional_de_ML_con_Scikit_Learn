# Tu primer árbol de decisión con scikit-learn

```python
#Importamos las librerias principales
import pandas as pd
import matplotlib.pyplot as plt
```

Utilizaremos el **dataset Titanic** de Standford: https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/problem12.html

```python
#Cargamos dataset a utilizar
#titanic = pd.read_csv("C:/Users/layla/Escritorio/Datasets/titanic.csv", sep = ",")
titanic = pd.read_csv('https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv')
```

## Exploración de datos

```python
#Visualizacion del DataFrame
titanic.head(10)
```

```python
titanic.count()
```

```python
#Eliminamos las columnas que no nos interesan
columns_to_drop = ['Name', 'Fare']
titanic.drop(columns_to_drop, axis = "columns", inplace = True)
```

```python
#Renombramos las columnas
titanic.columns = ['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch']
titanic.head()
```

```python
titanic.info()
```

```python
#Analizamos el shape del objeto
print(titanic.shape)
```

```python
#Visualizamos los tipos de datos
titanic.dtypes
```

```python
#Cambiamos los tipos de datos
titanic = pd.get_dummies(titanic, columns = ["Sex"], drop_first = True)
```

```python
titanic.dtypes
```

```python
titanic.head()
```

```python
#Reemplazar el DataFrame
titanic.rename(columns = {'Sex_male': 'Sex'}, inplace = True)
```

```python
#Ordenar columnas por nombres
titanic = titanic[['Survived','Pclass', 'Sex',  'Age', 'SibSp', 'Parch']]
titanic.head()
```

## Entrenamiento de árbol de decisión con scikit-learn

```python
#Separamos en X e y
X = titanic.drop("Survived", axis = 1)
y = titanic.Survived
```

```python
#Importamos las librerias necesarias para la creacion del modelo
from sklearn.model_selection import train_test_split

#30% para test y 70% para train
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size = 0.30, random_state = 00000)
```

```python
#Arbol de Decision
from sklearn.tree import DecisionTreeClassifier

#Creacion del modelo
tree = DecisionTreeClassifier(max_depth=2, random_state = 00000)
```

```python
#Entrenamiento
tree.fit(X_train,y_train)
```

## Evaluación del modelo

```python
#Calculo de las predicciones en Train y Test
y_train_pred = tree.predict(X_train)
y_test_pred = tree.predict(X_test)
```

```python
#Calculo de metricas
from sklearn.metrics import accuracy_score

#Calculo el accuracy en Train
train_accuracy = accuracy_score(y_train,y_train_pred)

#Calculo el accuracy en Test
test_accuracy = accuracy_score(y_test,y_test_pred)

print('El accuracy en train es:', train_accuracy)
print('El accuracy en test es:', test_accuracy)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

# Suponiendo que 'tree_classifier' es tu modelo de árbol de decisión
# ... (tu proceso de ajuste del modelo aquí)

# Ahora, para visualizar el árbol:
plt.figure(figsize=(20,10))  # Ajusta el tamaño según tus necesidades
tree.plot_tree(tree, filled=True)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier

# Creación del modelo con un nombre diferente
tree_classifier = DecisionTreeClassifier(max_depth=2, random_state=0)
# Entrenamiento
tree_classifier.fit(X_train, y_train)
# Cálculo de las predicciones en Train y Test
y_train_pred = tree_classifier.predict(X_train)
y_test_pred = tree_classifier.predict(X_test)
# Cálculo de métricas
from sklearn.metrics import accuracy_score

# Cálculo del accuracy en Train
train_accuracy = accuracy_score(y_train, y_train_pred)

# Cálculo del accuracy en Test
test_accuracy = accuracy_score(y_test, y_test_pred)

print('El accuracy en train es:', train_accuracy)
print('El accuracy en test es:', test_accuracy)

from sklearn import tree
import matplotlib.pyplot as plt

# Visualización del árbol:
plt.figure(figsize=(16, 8))  # Ajusta el tamaño según tus necesidades
tree.plot_tree(tree_classifier, filled=True)
plt.show()
```

```python
#Verificamos el feature importances
import seaborn as sns

importances = tree.feature_importances_
columns = X.columns
sns.barplot(x=columns, y=importances, palette = 'bright', saturation = 2.0, edgecolor = 'black', linewidth = 2)
plt.title('Importancia de cada Feature')
plt.show()
```
